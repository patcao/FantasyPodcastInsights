Simple Demo Script To Test Deepgram Transcription

```python
from deepgram import Deepgram
from pathlib import Path
import json
import os
from dotenv import load_dotenv

def transcribe_local_file(file_path, dg_client):
    """
    Transcribe a single local audio file using Deepgram
    Returns the transcript text and saves files
    """
    try:
        base_name = Path(file_path).stem
        print(f"Starting transcription of {base_name}...")

        # Transcribe the file
        with open(file_path, 'rb') as audio:
            source = {'buffer': audio, 'mimetype': 'audio/mpeg'}
            response = dg_client.transcription.sync_prerecorded(
                source,
                {
                    'smart_format': True,
                    'punctuate': True,
                    'model': 'nova-2',
                    'language': 'en-US',
                    'paragraphs': True  # Enable paragraph detection
                }
            )

        # Get the transcript text
        text_content = response['results']['channels'][0]['alternatives'][0]['transcript']

        print(f"\n✓ Successfully transcribed {base_name}")
        print("\nTranscription:")
        print("=" * 80)
        
        # Format the text into paragraphs
        # Split on sentences and create paragraphs
        sentences = text_content.split('. ')
        paragraphs = []
        current_paragraph = []
        
        for sentence in sentences:
            current_paragraph.append(sentence)
            if len(current_paragraph) >= 3:  # Adjust this number to control paragraph size
                paragraphs.append('. '.join(current_paragraph) + '.')
                current_paragraph = []
        
        # Add any remaining sentences
        if current_paragraph:
            paragraphs.append('. '.join(current_paragraph) + '.')
        
        # Print formatted paragraphs
        for paragraph in paragraphs:
            print(f"\n{paragraph}")
            
        print("\n" + "=" * 80)
        
        return text_content

    except Exception as e:
        print(f"Error transcribing {file_path}: {str(e)}")
        return None

def main():
    # Get API key from environment variable
    api_key = os.getenv("DEEPGRAM_API_KEY")
    
    # Initialize Deepgram client
    dg = Deepgram(api_key)

    # File path
    file_path = r"C:\Users\banee\Desktop\Columbia\LLM Driven NBA Fantasy Sleeper Prediction\github repos\FantasyPodcastInsights\data\raw\MP3 File\All-Star Weekend Takeaways Nets Deep Dive Schedule Notes  Fantasy Basketball QA.mp3"

    # Process the file
    transcribe_local_file(file_path, dg)

if __name__ == "__main__":
    main()
```
